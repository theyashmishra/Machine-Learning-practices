```python
from numpy import *
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
data
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
df['grad_rate']['95']=100
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('private',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
#making column 1 for yes 0 for no
def con(pri):
    if(pri=='Yes'):
        return 1
    else:
        return 0
```

```python
df['column']=df['private'].apply(con)
df['column']
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(confusion_matrix(df['column'],kmeans.labels_))
print(classification_report(df['column'],kmeans.labels_))
```

```python
df.drop('private',axis=1,inplace=True)
```

```python
pri=kmeans.predict(df.drop('column',axis=1))
```

```python
print(confusion_matrix(df['column'],pri))
print(classification_report(df['column'],pri))
```
